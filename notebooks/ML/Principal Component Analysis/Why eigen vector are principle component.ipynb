```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

```

<h1> Why eigen vector are principle component </h2>

```python
x = np.random.normal(0, 1, 700)
y = np.random.normal(0, 1, 700)
X = np.vstack((x, y)).T
S = np.array([[2,3],
              [1,0.3]])
X = S.dot(X.T).T
#X =  StandardScaler().fit(X).transform(X) ##centering of data
plt.scatter(X[:, 0], X[:, 1],color='b')

plt.title('Generated Data')
plt.axis('equal');
```

```python
S = np.cov(X.T)
S
```

```python
eigen_values , V = np.linalg.eig(S)
print("eigen values : " , eigen_values.T)
print("eigen vectors stacked V : " )
print(V)
```

```python
ax = plt.gca()
ax.arrow(0,0,eigen_values[0]*V[0,0],eigen_values[0]*V[1,0], head_width=0.1, head_length=1, fc='red', ec='black')
ax.arrow(0,0,eigen_values[1]*V[0,1],eigen_values[1]*V[1,1], head_width=0.1, head_length=1, fc='red', ec='black')
plt.scatter(X[:, 0], X[:, 1],color='b')
plt.axis('equal');
```

```python
new_X = X.dot(np.linalg.inv(V.T))
plt.scatter(new_X[:, 0], new_X[:, 1],color='b')
plt.axis('equal');
```

```python
new_X
ax = plt.gca()
S = np.cov(new_X.T)
eigen_values , V = np.linalg.eig(S)
print("eigen values : " , eigen_values.T)
print("eigen vectors stacked V : " )
print(V)
ax.arrow(0,0,eigen_values[0]*V[0,0],eigen_values[0]*V[1,0], head_width=0.1, head_length=1, fc='red', ec='black')
ax.arrow(0,0,eigen_values[1]*V[0,1],eigen_values[1]*V[1,1], head_width=0.1, head_length=1, fc='red', ec='black')
plt.scatter(new_X[:, 0], new_X[:, 1],color='b')
plt.axis('equal');
```
