```python
#### SOME IMPORT FUNCTION FOR PLOTING AND CALCULATING DERIVATE

def plot_function(x_args,y_args,show=True,label='',x_label='',y_label=''): 
    """
    takes x_args.y_args
    """
    plt.plot(x_args,y_args,label=label)
    plt.legend()
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    if show:
        plt.show()

def plot_function_f(f,show=True,label='',x_label='',y_label='',x_range=20,y_range=-20): 
    """
    takes the python function or lambda and plot it can be set up the range of function by para,eters x_range 
    and y_range
    """
    x = np.linspace(-x_range,x_range)
    y = [f(i) for i in x]
    plt.plot(x,y,label=label)
    plt.legend()
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    if show:
        plt.show()        


def plt_properties():
    """
    seting some propertios of plot object
    """
    ax = plt.gca()
    plt.xlim(-10, 10)
    plt.ylim(-10, 100)

    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['right'].set_color('none')
def calculate_derivative_at_point(function, point, precision = 1e-7):
    """
    Calculates a numerical approximation to the derivative of the specified function
    at the given point
    """
    der = (function(point + precision) - function(point)) / precision
    return der

def plot_derivative_at_point(function, point, derivative = None, min_x = -10, max_x = 10):
    """
    Plots the function in the range [x_min; x_max]. Computes the tangent line to the function
    at the given point and also plots it
    """
    plt_properties()
    vectorized_function = np.vectorize(function)

    x = np.linspace(min_x, max_x, 1000)
    y = vectorized_function(x)

    slope = 0  # Slope of the tangent line
    if derivative is None:
        slope = calculate_derivative_at_point(function, point)
    else:
        slope = derivative(point)

    intercept = function(point) - slope * point
    tangent_line_x = np.linspace(point - 4.5, point + 4.5, 10)
    tangent_line_y = slope * tangent_line_x + intercept
    plt.plot(x, y)
    plt.plot(tangent_line_x, tangent_line_y,color='r')        
    
    
def plot3D(f,label=''):
    """
    plot function with free arguments
    """
    fig = plt.figure(figsize=(20,20))
    ax = fig.add_subplot(111, projection='3d')
    x = np.linspace(-50,50)
    y = np.linspace(-50,50)
    
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)  
    #.plot3D(X, Y, Z, 'gray')      
    ax.contour3D(X, Y, Z, 50, cmap='binary')
    ax.set_xlabel('a',fontsize=25)
    ax.set_ylabel('b',fontsize=25)
    ax.set_zlabel('cost',fontsize=25)
    plt.legend()
```

```python
y = lambda x : x**2
x_args = np.linspace(-10,10)
y_args = y(x_args)
plot_function(x_args,y_args,label="$y=x^2$",x_label='x',y_label='y')
```

##### The above graphic immediatly shows  the min value of y  is  when  x = 0, but we have to make an algorithm wich is able to find numericaly x value ,especialy in more sophisticated functions which contains  many variables. 




#### When we give the different values on x the  f(x)= y   increase or decrease .Our goal is to give such values on x that y to tends to its min value using only its derivate   y' or in many variable the its  gradient .

#### Let's to dive in gradient descent
#### !!!! Remember !!!  We not interested in what is the minimum value   of y the  purpose is to find the value x where y gain or tends to its  min value

##### Let's to give the random variable for x ,for example x = -9 then we have f(-9) = 81 after  that  we have to make a choice what  is the increase  step  'h'  must to be in this way to achieve  f(-9+h)<f(-9). !!! As a whole we have to choice the sign of h (positive or negative)
##### Let  to be h=0.1 then we have 1) $\frac{y(-9+0.1) - f(-9)}{0.1}<0$ we can give for h = some positive number 0.1 or 0.5 and  y will always decrease,if we give h =-0.1 we get 2) $\frac{y(-9-0.1) - f(-9)}{0.1}>0$ this mean y increase .From 1) and 2)    we can see  that the h must to be  positive in case $\frac{y(-9+h) - f(-9)}{h}<0$ from MA wi know  $\frac{y(-9+h) - f(-9)}{h}=f'(9) + O(h) <0$  this brings us to idea ,the h must to be with oposite sign of sign of  derivate ,we can check that conclusion when   x=9 then   $\frac{y(9+h) - f(9)}{0.1}>0$ and f'(9)=18  The sighn of f'(9) >0 and h <0 ,We see that if we get h with sign opposite of first derivate the function decreace It can be proof it   with Taylor's series.The strong mathematical proof can be seen <a href="https://www.cs.princeton.edu/courses/archive/fall18/cos597G/lecnotes/lecture3.pdf">HERE</a> 



### Acording above expanations our algorithm when  give  the next value for x  must to required the folowing equation  $x1 - x  = h  = - \frac{f'(x)}{|f(x)|}*e$ where e is the learning rate   or  $x1 = x - - \frac{f'(x)}{|f(x)|}*e$



```python
def perform_gradient_descent(gradient,init_x=10,learning_step=0.1,count_iteration=1000):
    """
    calculates the gradient descent for Function of one variable
    """
    x_args = []
    y_args = []
    x = init_x
    x_last = 0
    for i in range(count_iteration):
        x = x - gradient(x)*learning_step
        x_args.append(i)
        y_args.append(x)
        x_last = x
        
        
    return round(x_last),x_args,y_args
```

```python
f_prim = lambda x :x*2
#result = perform_gradien_descent(f_prim,count_iteration=1000)
x_last,x_args,y_args = perform_gradient_descent(f_prim)
plot_function(x_args,y_args,x_label='iteration',y_label='x')
print("x-->",x_last)
```

##### this algoritims can be apply for the function of many variable

```python
def gradient_desent(gradient_x,gradient_y,init_x=-10,init_y=-10,learning_step=0.1,count_iteration=1000):
    """
    calculates the gradient descents of many variable
    """
    x = init_x
    y= init_y
    for i in range(count_iteration):
        new_x = x - gradient_x(x,y)*learning_step
        new_y = y - gradient_y(x,y)*learning_step
        x = new_x
        y = new_y
    
    print('(x,y)=', math.ceil(x),math.ceil(y) )
    return math.ceil(x),math.ceil(y)  


            
```
